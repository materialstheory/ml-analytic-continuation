```python
import numpy as np
from scipy.integrate import simps
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
A_omega_test = []
```

triangular\
$A(\omega) = \frac{1}{32}(\omega + 4)$
with $ -4< \omega \leq 4$

```python
omega = np.linspace(-8, 8, 800)
A = np.zeros(800)
A[200: 600] = (omega[200:600] + 4)/32
plt.plot(omega, A)
A_omega_test.append(A)
```

rectangular\
$A(\omega) = \frac{1}{8}$ 

```python
A = np.zeros(800)
A[200:600] = 1/8

plt.plot(omega, A)
A_omega_test.append(A)
```

$\sigma = 1 \quad r = 2$\
Asymmetric peak/plateau \
$A(\omega) = \frac{1}{\sigma \sqrt{2\pi}} \exp({-\frac{1}{2}{(\frac{(\sinh(\omega - r)- 4)\times 0.1}{\sigma})}^{2}})$


```python
sigma = 1
r = 2
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * ((np.sinh(omega- r) -4)*0.1/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```

plateau\
$A(\omega) = \frac{1}{\sigma\sqrt{2\pi}}\exp{(-\frac{1}{2}{(\frac{(\cosh(\omega - r)- 4)\times 0.1}{\sigma})}^{2})}$

```python
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * ((np.cosh(omega- r) -4)*0.1/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```

assymetric peaks/plateau\
$A(\omega) = \frac{1}{\sigma\sqrt{2\pi}}\exp{(-\frac{1}{2}{(\frac{((\omega - r)^{3} - 4)\times 0.1}{\sigma})}^{2})}$

```python
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * (((omega- r)**3 -4)*0.1/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```

assymetric peak\
$A(\omega) = \frac{1}{\sigma\sqrt{2\pi}}\exp{(-\frac{1}{2}(\frac{(\sinh(\omega - r) - 4)\times 0.6}{\sigma})^{2})}$

```python
sigma = 1
r = 2
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * ((np.sinh(omega- r) -4)*0.6/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```

Assymetric peak at the center
$A(\omega) = \frac{1}{\sigma\sqrt{2\pi}}\exp{(-\frac{1}{2}(\frac{(\sinh(\omega + 1) - 4)\times 0.6}{\sigma})^{2})}$

```python
sigma = 1
r = 2
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * ((np.sinh(omega + 1) -4)*0.6/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```

$A(\omega) = \frac{1}{\sigma\sqrt{2\pi}}\exp{(-\frac{1}{2}(\frac{(\sinh(\omega + 1) - 2)\times 0.6}{\sigma})^{2})}$

```python
sigma = 1
r = 2
A = 1/sigma/np.sqrt(2*np.pi) *  np.exp(-0.5 * ((np.sinh(omega + 1) -2)*0.6/sigma)**2)
plt.plot(omega, A/simps(A, omega))
A_omega_test.append(A/simps(A, omega))
```
